```python
# First ascending and then descinding
df.sort_values(by=['Name','Type 1'],ascending=[1,0])
```

```python
# Adding the numeric values
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 
```

```python
df.head()
```

```python
# To drop a specific column in the dataFrame
df.drop(['Total','Generation'],axis=1)
```

```python
df = df.drop(['Total'],axis=1)
```

```python
# To add the column [rows:rows,columns:columns]
df['Total'] = df.iloc[:,4:10].sum(axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['#', 'Name', 'Type 1', 'Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk',
       'Sp. Def', 'Speed', 'Total','Generation', 'Legendary', ]]
```

```python
df.head()
```

```python
# To save the file
df.to_csv('pokemon_modified.csv',index=False)
```

```python
df = pd.read_csv('pokemon_modified.csv')
```

```python
## Getting top 10 Pokemon Total
df.groupby('Name').sum()['Total'].sort_values(ascending=False)[1:10]
```

```python
df.groupby('Type 1').mean()['Total'].sort_values(ascending=False)[1:10]
```

```python
# Group by two columns
df.groupby(['Type 1','Type 2']).mean()['Total'].sort_values(ascending=False)[1:10]
```
